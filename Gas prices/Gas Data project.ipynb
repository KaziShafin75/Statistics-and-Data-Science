```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
gas = pd.read_csv("E:\Matplotlib\gas_prices.csv")
gas
```

```python
gas= gas.drop(gas.index[0])
```

```python
gas.head(5)
```

```python
gas['asia']= gas['Australia']+gas['Japan']+gas['South Korea']
gas['europe']= gas["France"]+gas["Germany"]+gas["Italy"]+gas["UK"]
gas['america']= gas["USA"]+gas["Mexico"]+gas["Canada"]
```

```python
gas
```

#Gas price distribution

```python
plt.hist(gas['asia'],bins=6,density=True,color='magenta',alpha=0.6,edgecolor="black")
plt.title("Gas distribution of Asian countries")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(gas['europe'],bins=6,density=True,color='red',alpha=0.6,edgecolor="black")
plt.title("Gas distribution of European countries")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(gas['america'],bins=6,density=True,color='blue',alpha=0.6,edgecolor="black")

plt.title("Gas distribution of American countries")

plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

#Comparison of gas prices with different graphs

```python
plt.figure(figsize=(8,5))
plt.plot(gas['Year'],gas['Australia'],'r.-',label='Australia')
plt.plot(gas['Year'],gas['Japan'],'g.-',label='Japan')
plt.plot(gas['Year'],gas['South Korea'],'b.-',label='South Korea')
plt.title("Gas Prices in Asia (in USD)")
plt.xticks(gas['Year'],rotation='vertical')
plt.xlabel("Year")
plt.ylabel("Gas Price")
plt.legend()
plt.show()
```

```python
x= gas['Year']
y=gas['europe']
plt.bar(x,y,color='g')
plt.title("Gas price in Europe in (USD)")
plt.xticks(gas['Year'],rotation='vertical')
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

```python
x=gas["america"][::2]
y=gas['Year'][::2]
explode=(0,0.2,0.2,0,0,0,0,0,0)
plt.title("Gas price in American Countries in different years")
plt.pie(x,labels=y,autopct="%1.1f %%",explode=explode)
plt.show()
```

#scatterplot

```python
plt.style.use('default')
plt.figure(figsize=(5,7))
x= gas['asia']
y=gas['europe']
z=gas['america']
labels=["Asia","Europe","America"]
plt.title("Gas price comparison of different regions")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.boxplot([x,y,z],tick_labels=labels)
plt.show()
```

#Correlation

```python
x= gas['asia']
y=gas['europe']
z=gas['Year']
fig1,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(z,x,color='g',alpha=0.6)
ax2.plot(z,y,'r-')
ax1.set_ylabel("Asia's gas price over the years")
ax2.set_ylabel("Eupore's gas price over the years")

plt.show()
```

#Correlation os Asia & America

```python
x=gas['asia']
y=gas['america']
plt.xlabel("Asia's gas price over the years")
plt.ylabel("Ameica's gas price over the years")
plt.title("Correlation between Asia and America")
plt.scatter(x,y,color='blue',alpha=0.9,marker='^')
plt.show()
```

```python
plt.hist(gas['europe'],bins=3,color='#d60d75')
plt.hist(gas['america'],bins=5,color="#329984")
plt.title("Correlation between Europe and America")
plt.xlabel("Price intervals")
plt.ylabel("Frequency")
plt.show()
```
