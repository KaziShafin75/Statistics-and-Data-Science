# Now we will see the distributions of s,p ; lengths and widths and also the rltn... between them

```python
s_l=df[df.columns[0]].astype(float)
s_w= df[df.columns[1]].astype(float)
p_l=df[df.columns[2]].astype(float)
p_w=df[df.columns[3]].astype(float)
```

```python
s_l.describe()
```

```python
plt.title("Distribution of sepal_length")

plt.hist(s_l,density=True,alpha=0.7,color="blue")
mu= np.mean(s_l)
sigma=np.std(s_l)
x=np.linspace(4,8,1000)
pdf=stats.norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='r',linewidth=2)
plt.xlabel("sepal-length")
plt.ylabel("Density")
plt.text(9,0.3,"The distribution is approximately normal!",fontsize=20)

plt.show()
```

```python
p_w.describe()
```

```python
plt.title("Distribution of petal_width",color="b")
plt.hist(p_w,density=True,color="#f20266")
mu= np.mean(p_w)
sigma=np.std(p_w)
x=np.linspace(0,3,1000)
pdf=stats.norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='b')
plt.xlabel("petal_width")
plt.ylabel("Density")
plt.text(3.5,0.5,"The distribution is not normal!",fontsize=20)

plt.show()
```

```python
plt.hist(s_w)
plt.title("Distribution of sepal_width")
plt.xlabel("sepal-width")
plt.ylabel("Density")

plt.show()
```

```python
cor_s = np.corrcoef(s_l,s_w)
print(cor_s[0,1])
plt.scatter(s_l,s_w,marker=".")
plt.show()
# The relation is not too strong and it is also negetive, if sepal_length increses, sepal_width decreses
```

# Boxplot

```python
plt.style.use('default')
plt.figure(figsize=(5,7))

setosa= df.loc[df['species']=="Iris-setosa"]['sepal_length']
versicolor=df.loc[df['species']=='Iris-versicolor']['sepal_length']
virginica=df.loc[df['species']=='Iris-virginica']['sepal_length']

plt.title("Boxplot of Different type of Iris flowes classfied according to sepal_length")

labels= ["Iris-setosa","Iris-versicolor", "Iris-virginica"]
boxes=plt.boxplot([setosa,versicolor,virginica],tick_labels=labels,patch_artist=True)

for box in boxes['boxes']:
    box.set(facecolor='green', color='#1182a8', linewidth=2)
plt.text(4,6,"There's an outlier in Iris-virginica",fontsize=20)    
plt.show()
```
